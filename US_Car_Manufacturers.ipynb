## **Top 5 car brands in terms of number of cars**

```python
data = []

for i in df['brand'].unique():
  data.append((i, len(df[df['brand'] == i])))
```

```python
data
```

```python
#Converting data list into dataframe
data = pd.DataFrame(data, columns = ['brand', 'frequency'])
```

```python
data
```

```python
#Arranging data in sorted form
data.sort_values(by = 'frequency', ascending = False)
```

```python
#Top 5 car brands
data = data.sort_values(by = 'frequency', ascending = False).head()
```

```python
data
```

```python
#Plotting the graph for the above data
data.plot(x = 'brand', y = 'frequency', kind = 'bar', figsize = (5,5))
```

### **Car Industry Trend**

```python
data1 = []

for j in df['year'].unique():
  data1.append((j, len(df[df['year'] == j])))
```

```python
data1 = pd.DataFrame(data1, columns = ['year', 'frequency'])
```

```python
data1 = data1.sort_values(by = 'year', ascending = True)
```

```python
data1
```

```python
data1.plot(x = 'year', y = 'frequency', kind = 'line', figsize = (10,5))
```

### **Most popular car colour**

```python
data2 = []

for k in df['color'].unique():
  data2.append((k, len(df[df['color'] == k])))

data2 = pd.DataFrame(data2, columns = ['color', 'frequency'])

data2 = data2.sort_values(by = 'frequency', ascending = False)

data2.plot(x = 'color', y = 'frequency', kind = 'bar', figsize = (15,5))
```

### **Which car brand has maximum average price**

```python
df.groupby('brand').mean()['price'].sort_values(ascending = False).plot.bar(figsize = (15,5))
```

### **Which state has the most expensive cars** 

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (15,5))
df.groupby('state').mean()['price'].sort_values(ascending = False).plot.bar(color = 'orange')
plt.show()
```
